```python
!pip install nltk
import nltk
nltk.download("all")
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
words=word_tokenize(text)
print(len(words))
print(words)
```

```python
sentences=sent_tokenize(text)
print(len(sentences))
print(sentences)
```

```python
from nltk.probability import FreqDist
fdist=FreqDist(words)
fdist.most_common(10)
```

```python
import matplotlib.pyplot as plt
fdist.plot(10)
```

```python
words_no_punc=[]
for w in words:
  if w.isalpha():
    words_no_punc.append(w.lower())
```

```python
print(words_no_punc)
print("\n")
print(len(words_no_punc))
```

```python
fdist=FreqDist(words_no_punc)
fdist.most_common(10)
fdist.plot(10)
```

```python
from nltk.corpus import stopwords
stopwords=stopwords.words("english")
print(stopwords)
```

```python
clean_words=[]
for w in words_no_punc:
  if w not in stopwords:
    clean_words.append(w)
```

```python
print(clean_words)
print("\n")
print(len(clean_words))
```

```python
fdist=FreqDist(clean_words)
fdist.most_common(10)
fdist.plot(10)
```

```python
from nltk.corpus import wordnet
nltk.download('wordnet')
for word in wordnet.synsets("Games"):
  print(word)
```

```python
for words in wordnet.synsets("Fun"):
  for lemma in words.lemmas():
    print(lemma)
print("\n")
```

```python
word= wordnet.synsets("Play")[0]
print(word.name)

print(word.definition())

print(word.examples())
```

```python
for words in wordnet.synsets("Fun"):
  print(words.name())
  print(words.definition())
  print(words.examples())
  for lemma in words.lemmas():
    print(lemma)
print("\n")
```

```python
word= wordnet.synsets("Play")[0]
print(word.hypernyms())
```

```python
word= wordnet.synsets("Play")[0]
print(word.hyponyms())
```

```python
synonyms=[]
antonyms=[]
for words in wordnet.synsets('good'):
  for lemma in words.lemmas():
    synonyms.append(lemma.name())
    if lemma.antonyms():
      antonyms.append(lemma.antonyms()[0].name())
print(synonyms)
print("\n")
print(antonyms)
```

```python
synonyms=[]
antonyms=[]
for words in wordnet.synsets('New'):
  for lemma in words.lemmas():
    synonyms.append(lemma.name())
    if lemma.antonyms():
      antonyms.append(lemma.antonyms())
print(synonyms)
print("\n")
print(antonyms)
```
